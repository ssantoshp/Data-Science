```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv(r"C:\Users\Geetha\Downloads\india.csv")
df.drop(['Area_Name','Year','Subgroup','Rape_Cases_Reported','Victims_of_Rape_Total'],axis=1,inplace=True)
```

```python
df
```

```python
 a = df['Victims_Above_50_Yrs'].sum()
print (a)

```

```python
 b = df['Victims_Between_10-14_Yrs'].sum()
print (b)
```

```python
 c = df['Victims_Between_14-18_Yrs'].sum()
print (c)
```

```python
 d = df['Victims_Between_18-30_Yrs'].sum()
print (d)
```

```python
 e = df['Victims_Between_30-50_Yrs'].sum()
print (e)
```

```python
 f = df['Victims_Upto_10_Yrs'].sum()
print (f)
```

```python
plt.title('Number of rape cases in India since 2001 to 2010 depending on the age of the victim')

labels=['-10','10-14','14-18','18-30','30-50','+50']
values=[f,b,c,d,e,a]
plt.ylabel('Number of rape cases')

plt.bar(labels, values)
plt.show()
```
