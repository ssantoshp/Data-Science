```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label = '2x', color='red')

plt.title('A graph showing nothing')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
x2= np.arange(0,4.5,0.5)
plt.figure(figsize=(5,3),dpi=150)
plt.plot(x2,x2**2, label= '2x**2')

plt.plot(x,y, label ='2x')

plt.title('A graph showing nothing')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.legend()


plt.show()
```
