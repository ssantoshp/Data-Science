```python
A = s.str.count('Administration')
B = s.str.count('Anthropology')
C = s.str.count('Art')
D = s.str.count('Chemical')
E = s.str.count('Computer Engineering')
F = s.str.count('Computer Science')
G = s.str.count('Economics')
H = s.str.count('Environmentally')
I = s.str.count('Finance')
J = s.str.count('Law')
K = s.str.count('MBA')
L = s.str.count('Management')
M = s.str.count('Marketing')
N = s.str.count('Math')
O = s.str.count('Med')
P = s.str.count('Nuclear')
Q = s.str.count('Philo')
R = s.str.count('Physics')
S = s.str.count('Electrical Engineering')
T = s.str.count('Poli')
U = s.str.count('Psycho')
V = s.str.count('English')

```

```python
majors = {'Administration': A.sum(), 'Anthropology': B.sum(), 'Art':C.sum(), 'Chemical Engineering':D.sum(),'Computer Engineering':E.sum(),'Computer Science':F.sum(),'Economics':G.sum(),'Environmentally sustainable business':H.sum(),'Finance':I.sum(),'Law': J.sum(),"MBA":K.sum(),'Management':L.sum(),"Marketing":M.sum(),'Mathematics':N.sum(),'Medicine':O.sum(),'Nuclear Engineering':P.sum(),'Philosphy':Q.sum(),'Physics':R.sum(),'Political Science':T.sum(), 'Psycholgy':U.sum(),'Electrical Engineering':S.sum(),'English':V.sum()}
major = pd.Series(majors)
```
